```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.loadtxt('stuff.csv', [('a1x', 'f4'), ('a1y', 'f4'), ('a1z', 'f4'),
    ('g1x', 'f4'), ('g1y', 'f4'), ('g1z', 'f4')], delimiter=',')

a1x = x['a1x'] / 16384
a1y = x['a1y'] / 16384
a1z = x['a1z'] / 16384
a1 = a1x * a1x + a1y * a1y + a1z * a1z

time_step = 5 / 1000
freqs = freqs = np.fft.fftfreq(a1x.size, time_step)
idx = np.argsort(freqs)
```

```python
plt.figure(figsize=(20, 10))
plt.plot([x * time_step for x in range(a1.size)], a1)
```

```python
a1x_spectrum = np.abs(np.fft.fft(a1x))
plt.plot(freqs[idx], a1x_spectrum[idx])
```

```python
a1y_spectrum = np.abs(np.fft.fft(a1y))
plt.plot(freqs[idx], a1y_spectrum[idx])
```

```python
a1z_spectrum = np.abs(np.fft.fft(a1z))
plt.plot(freqs[idx], a1z_spectrum[idx])
```
